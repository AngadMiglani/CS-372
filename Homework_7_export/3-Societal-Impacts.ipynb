# Part 3: Societal Impacts

There is no programming in this part of the assignment. Instead, this part is about reading a paper on the societal impacts and ethical considerations of generative image models. You will choose a paper of interest and respond to a reflection question.

### Learning Objectives

Through this exercise, you will:
1. **Read** about the societal impacts and ethical considerations (including demographic bias, harm to artists, and deepfakes) of generative image models.
2. **Interpret** the possible societal risks and **articulate** your own view in response to critical reflection questions.

To begin, choose at least one of the following three articles to read:
1. [*Easily Accessible Text-to-Image Generation Amplifies Demographic Stereotypes at Large Scale*](https://dl.acm.org/doi/10.1145/3593013.3594095) by Bianchi et al., ACM FAccT 2023
2. [*AI Art and its Impact on Artists*](https://dl.acm.org/doi/10.1145/3600211.3604681) by Jiang et al., ACM/AAAI AIES 2023
3. [*Deepfakes on Demand: The rise of accessible non-consensual deepfake image generators*](https://dl.acm.org/doi/10.1145/3715275.3732107) by Hawkins et al., ACM FAccT 2025

## Task 1

Referencing the article but in your own words, answer one of the following critical reflection questions (corresponding to the article you chose). Explain in 2-3 paragraphs. Make sure to clarify which article you chose / which prompt you are responding to.

1. Bianchi et al. find that text-to-image models amplify rather than merely reflect stereotypes and societal biases. What does it mean for a model to amplify rather than reflect stereotypes, and why might this distinction be important for image generation models? Moreover, they show that neither prompting nor platform-level guardrails successfully prevent such stereotype amplification, and ultimately conclude "We urge users to exercise caution and refrain from using such image generation models in any applications that have downstream effects on the real-world..." -- Based on the evidence presented, do you agree with this recommendation? Why or why not?

2. Jiang et al. document a range of harms that generative AI image tools may cause for artists, including economic displacement, plagiarism, reputational damage, and the use of artists' work as training data without consent. Which of these harms do you find most serious, and why? In the end, do you believe artists have a right to control how their creative work is used in training AI systems? If so, what is the basis of that right? If not, what more limited rights, if any, do you believe artists should enjoy?

3. Hawkins et al. present empirical evidence that nearly 35,000 publicly downloadable deepfake model variants exist on popular repositories, downloaded almost 15 million times, with 96% targeting women and many signaling intent to generate non-consensual intimate imagery. These models can be created with as few as 20 images and 15 minutes on a consumer-grade computer. Reflect on the scale and accessibility of this problem. The authors note that these models violate the platforms' own Terms of Service and in some cases existing law, yet they persist. What does this gap between policy and practice tell us about the adequacy of current approaches to governing AI harms? What kinds of interventions — technical, legal, social, or otherwise — do you think could meaningfully address this issue?

Answer 2:

Jiang et al. identify several harms associated with generative image models, including economic displacement, plagiarism, reputational risks, and the use of artists’ work as training data without consent. Among these, non-consensual data use appears particularly significant because it underlies many of the other concerns. When models are trained on datasets that include artists’ work without permission, they can reproduce stylistic elements that reflect accumulated creative labor, raising questions about how ownership is defined when outputs are derived from large, aggregated datasets. This issue also reflects broader patterns of value extraction in digital economies, where individual contributions are incorporated into scalable systems without clear mechanisms for recognition or compensation. In this context, artists may have a reasonable claim to some degree of control over how their work is used in training AI systems, grounded in existing intellectual property principles that aim to protect creative labor. However, the scale and technical nature of machine learning complicate direct application of these frameworks. As a result, approaches such as opt-out systems or collective licensing may offer a more practical balance between supporting innovation and addressing concerns about fairness, though these solutions are unlikely to fully resolve the underlying tensions.
